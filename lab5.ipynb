
##<b>Task 1<br></b>
Given a <b>$\mathbb{R}=1$ </b>vector in a 2-dimensional plane and visualize it using the cells below. In the first markdown cell below, write your vector using LaTeX. You should represent it as a system of linear equations. 

## <b>Rank 1 Vector: </b>
$$X=\begin{bmatrix}2\\8\\\end{bmatrix}$$

##<b>Linear Equation:</b>
$$ X=2x+8y$$

```
X=np.array([1,7])
print('the rank of vector X = ',np.linalg.matrix_rank(X))
```

```
#chart 
fig_1=plt.figure(1,figsize=(15,10))
chart_1=fig_1.add_subplot(221)
plt.xlim(0,10,)
plt.ylim(0,10)
plt.axhline(y=0, color='k',)
plt.axvline(x=0, color='k')
chart_1.set_ylabel('Y-Axis',fontsize='14')
chart_1.set_xlabel('X-Axis',fontsize='14')
plt.title('Plot of a Rank 1 Vector', fontweight='bold',fontsize='18',)

#plot of vector X
plt.quiver(0,0, X[0], X[1], angles='xy', scale_units='xy',scale=1, color='orange')
#plt.quiver([0,0],[0,0], X[0], X[1],  angles='xy', scale_units='xy',scale=1,color=['yellow'])

#display
plt.grid(color='blue')
plt.show()
```

##<b>Task 2</b>
Given a $\mathbb{R}=2$ vector in a 3-dimensional plane and visualize it using the cells below. In the first markdown cell below, write your vector using LaTeX. You should represent it as a system of linear equations.

## <b>Rank 2 Vector:</b>
$$array=\begin{Bmatrix}X=\begin{bmatrix} 1\\5\\4 \\\end{bmatrix},\;\:Y= \begin{bmatrix}4\\-2\\3\\\end{bmatrix}\end{Bmatrix}$$
##<b>Linear Equation:</b>
$$X=x+5y+4z\;\;\;\:Y=4x-2y+3z$$


```
K=np.array([
            [1,5,4],
            [4,-2,3]
])

print('the rank of vector = ',np.linalg.matrix_rank(K))
```

```
#figure

fig= plt.figure(figsize=(10,10))
ax= fig.gca(projection='3d')
ax.set_xlim([0,9])
ax.set_ylim([0,9])
ax.set_zlim([0,9])

ax.quiver(0, 0, 0, K[0,0], K[0,1],K[0,2], arrow_length_ratio=0.05,color='orange',label='X')
ax.quiver(0, 0, 0, K[1,0], K[1,1],K[1,2], arrow_length_ratio=0.05,color='blue',label='Y')

plt.legend()
 
```
